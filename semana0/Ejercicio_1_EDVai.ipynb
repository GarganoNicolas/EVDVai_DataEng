```python
# Lee el archivo Excel, leo los datos de cada hoja del archivo excel

excel_file = pd.ExcelFile('Utimas_Desvinculaciones.xlsx')

# Crea un diccionario para almacenar los DataFrames de cada hoja

dfs = {}

# Itera sobre cada hoja y guarda los datos en el diccionario 'dfs'

for sheet_name in excel_file.sheet_names:
    dfs[sheet_name] = excel_file.parse(sheet_name)

# Accede a las hojas específicas y asígnalas a variables individuales

data = dfs['Row data']
df_Rangos = dfs['Rangos']
df_Managers = dfs['Managers']

# Ahora puedes usar 'data', 'df_Rangos' y 'df_Managers' como variables independientes
```

### 2.4 - Limpieza y transformación de los datos

```python
# Convierte la columna "Fecha comienzo" a tipo datetime (CAMBIO DE FORMATO)

data['Fecha comienzo'] = pd.to_datetime(data['Fecha comienzo'], errors='coerce')

# Formatea la columna "Fecha comienzo" con el nuevo formato "dia-mes-año"
data['Fecha comienzo'] = data['Fecha comienzo'].dt.strftime('%d-%m-%Y')
```

```python
# Convierte la columna "Fecha fin" a tipo datetime (CAMBIO DE FORMATO)

data['Fecha fin'] = pd.to_datetime(data['Fecha fin'], errors='coerce')

# Formatea la columna "Fecha fin" con el nuevo formato "dia-mes-año"
data['Fecha fin'] = data['Fecha fin'].dt.strftime('%d-%m-%Y')
```

# 3 - Analisis de Datos

### 3.1 Primer analisis

```python
data.head()
```

**Cantidad de personas desvinculadas por año**

```python
# Extraer el año de la fecha de fin
data['Año'] = data['Fecha fin'].dt.year

# Contar la cantidad de personas por año
conteo_por_año = data.groupby('Año').size()

# Mostrar los resultados
print(conteo_por_año)

# Crear el gráfico de barras
plt.bar(conteo_por_año.index, conteo_por_año.values)

# Agregar etiquetas y título
plt.xlabel('Año')
plt.ylabel('Cantidad de personas')
plt.title('Cantidad de personas Desvinculadas por año')

# Mostrar el gráfico
plt.show()
```

**Cantidad de personas por Manager**

```python
# Contar la cantidad de personas por manager
conteo_por_manager = data['Manager'].value_counts()

# Graficar el gráfico de barras
conteo_por_manager.plot(kind='bar')

# Agregar etiquetas y título
plt.xlabel('Manager')
plt.ylabel('Cantidad de personas')
plt.title('Cantidad de personas por manager')

# Mostrar el gráfico
plt.show()
```

**Personas que Realizaron y No Realizaron Cursos**

```python
# Contar la cantidad de personas que realizaron cursos y las que no
realizo_cursos = sum(1 for curso in data['Realizo Cursos'] if curso == 'SI')
no_realizo_cursos = sum(1 for curso in data['Realizo Cursos'] if curso == 'NO')

# Mostrar los datos
print("Cantidad de personas que realizaron cursos:", realizo_cursos)
print("Cantidad de personas que no realizaron cursos:", no_realizo_cursos)

# Graficar
fig, ax = plt.subplots()
ax.pie([realizo_cursos, no_realizo_cursos], labels=['Realizó Curso', 'No Realizó Curso'], autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Proporción de Personas que Realizaron y No Realizaron Cursos')
plt.show()
```

```python
# Contar la cantidad de personas con rango salarial mayor que la competencia
mayor_que_competencia = sum(1 for rs, rs_competencia in zip(data['Rango Salarial'], data['RS competencia']) if rs > rs_competencia)

# Calcular la cantidad de personas con igual o menor rango salarial que la competencia
igual_o_menor_que_competencia = len(data['Rango Salarial']) - mayor_que_competencia

# Mostrar los datos
print("Cantidad de personas con rango salarial mayor que la competencia:", mayor_que_competencia)
print("Cantidad de personas con igual o menor rango salarial que la competencia:", igual_o_menor_que_competencia)

# Graficar
plt.bar(['Mayor que Competencia', 'Igual o Menor que Competencia'], [mayor_que_competencia, igual_o_menor_que_competencia])
plt.xlabel('Comparación')
plt.ylabel('Cantidad de Personas')
plt.title('Comparación del Rango Salarial con la Competencia')
plt.show()
```

```python
# Contar la cantidad de personas por área
conteo_por_area = {}
for area in data['Area']:
    if area not in conteo_por_area:
        conteo_por_area[area] = 0
    conteo_por_area[area] += 1

# Mostrar los datos
print("Cantidad de personas por área:")
for area, cantidad in conteo_por_area.items():
    print(f"{area}: {cantidad}")

# Graficar
plt.bar(conteo_por_area.keys(), conteo_por_area.values())
plt.xlabel('Área')
plt.ylabel('Cantidad de Personas')
plt.title('Cantidad de Personas por Área')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor visualización
plt.show()
```

### 3.2  Segundo Analisis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un gráfico de barras utilizando seaborn
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Grupo recruitment', hue='Area')
plt.title('Distribución de Grupo recruitment por Area')
plt.xlabel('Grupo recruitment')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para una mejor visualización
plt.legend(title='Area')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Orden deseado para el eje x
orden_grupo = ['Grupo C', 'Grupo A', 'Grupo B']

# Crear una copia del DataFrame data
data_ordenado = data.copy()

# Cambiar el orden de los niveles en la columna 'Grupo recruitment'
data_ordenado['Grupo recruitment'] = pd.Categorical(data_ordenado['Grupo recruitment'], categories=orden_grupo, ordered=True)

# Crear un gráfico de barras utilizando seaborn con el orden especificado
plt.figure(figsize=(10, 6))
sns.countplot(data=data_ordenado, x='Area', hue='Grupo recruitment')
plt.title('Distribución de Area por Grupo recruitment')
plt.xlabel('Area')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para una mejor visualización
plt.legend(title='Grupo recruitment', loc='upper right')
plt.tight_layout()
plt.show()
```
