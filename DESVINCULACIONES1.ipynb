El ratios nos brindan una comprensión de la distribución de desvinculaciones en cada área de la organización. Por ejemplo, áreas con ratios más altos pueden requerir una atención especial en términos de retención de empleados, gestión de recursos humanos y desarrollo de estrategias para mejorar la satisfacción y el compromiso del personal.
IT (Tecnologías de la Información): El ratio de desvinculaciones en el área de IT es del 26.67%. Esto significa que aproximadamente el 26.67% de todas las desvinculaciones ocurrieron en el área de Tecnologías de la Información.

```python
data['Realizó Cursos'] = (data['Realizo Cursos'] == 'SI').astype(int)

data['No Realizó Cursos'] = (data['Realizo Cursos'] == 'NO').astype(int)

conteo_cursos_por_area = data.groupby(['Area', 'Realizó Cursos']).size().unstack(fill_value=0)
conteo_cursos_por_area.columns = ['No Realizó Cursos', 'Realizó Cursos']

conteo_cursos_por_area['Total Empleados'] = conteo_cursos_por_area.sum(axis=1)
conteo_cursos_por_area['% Realizó Cursos'] = (conteo_cursos_por_area['Realizó Cursos'] / conteo_cursos_por_area['Total Empleados']) * 100
conteo_cursos_por_area['% No Realizó Cursos'] = (conteo_cursos_por_area['No Realizó Cursos'] / conteo_cursos_por_area['Total Empleados']) * 100

conteo_cursos_por_area = conteo_cursos_por_area.sort_values(by='Total Empleados', ascending=False)

plt.figure(figsize=(10, 6))

plt.bar(conteo_cursos_por_area.index, conteo_cursos_por_area['Realizó Cursos'], color='plum', label='SI')
plt.bar(conteo_cursos_por_area.index, conteo_cursos_por_area['No Realizó Cursos'], bottom=conteo_cursos_por_area['Realizó Cursos'], color='mediumorchid', label='NO')

for index, row in conteo_cursos_por_area.iterrows():
    plt.text(index, row['Realizó Cursos'] / 2, f"{row['% Realizó Cursos']:.1f}%", ha='center', va='center', color='white', fontweight='bold')
    plt.text(index, row['Realizó Cursos'] + row['No Realizó Cursos'] / 2, f"{row['% No Realizó Cursos']:.1f}%", ha='center', va='center', color='black', fontweight='bold')

plt.title('Número de Empleados que Realizaron y No Realizaron Cursos por Área')
plt.xlabel('Área')
plt.ylabel('Número de Empleados')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```
