# ⏰ DURACION

```python
#duración del empleo restando 'Fecha comienzo' de 'Fecha fin'
data['Fecha comienzo'] = pd.to_datetime(data['Fecha comienzo'])
data['Fecha fin'] = pd.to_datetime(data['Fecha fin'])
data['Duración del empleo'] = data['Fecha fin'] - data['Fecha comienzo']
print(data[['Fecha comienzo', 'Fecha fin', 'Duración del empleo']])
```

```python

import matplotlib.pyplot as plt
data['Fecha comienzo'] = pd.to_datetime(data['Fecha comienzo'])
data['Fecha fin'] = pd.to_datetime(data['Fecha fin'])
data['Duración del empleo'] = (data['Fecha fin'] - data['Fecha comienzo']).dt.days
plt.hist(data['Duración del empleo'], bins=20, edgecolor='black')
plt.title('Distribución de la duración del empleo')
plt.xlabel('Duración del empleo (días)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python

data['Fecha comienzo'] = pd.to_datetime(data['Fecha comienzo'])
data['Fecha fin'] = pd.to_datetime(data['Fecha fin'])
data['Duración del empleo'] = (data['Fecha fin'] - data['Fecha comienzo']).dt.days / 365
media = round(data['Duración del empleo'].mean(), 2)
mediana = round(data['Duración del empleo'].median(), 2)
maximo = round(data['Duración del empleo'].max(), 2)
minimo = round(data['Duración del empleo'].min(), 2)
print("Media de la duración del empleo:", media)
print("Mediana de la duración del empleo:", mediana)
print("Máximo de la duración del empleo:", maximo)
print("Mínimo de la duración del empleo:", minimo)
```

```python
# Calcular el promedio de la duración del empleo para cada área
data['Fecha comienzo'] = pd.to_datetime(data['Fecha comienzo'])
data['Fecha fin'] = pd.to_datetime(data['Fecha fin'])
data['Duración del empleo'] = (data['Fecha fin'] - data['Fecha comienzo']).dt.days / 365
promedio_por_area = data.groupby('Area')['Duración del empleo'].mean()
print("Promedio de tiempo de servicio por área:")
print(promedio_por_area)
```

```python
import matplotlib.pyplot as plt
data['Fecha comienzo'] = pd.to_datetime(data['Fecha comienzo'])
data['Fecha fin'] = pd.to_datetime(data['Fecha fin'])
data['Duración del empleo'] = (data['Fecha fin'] - data['Fecha comienzo']).dt.days / 365
promedio_por_area = data.groupby('Area')['Duración del empleo'].mean().sort_values(ascending=False)
ax = promedio_por_area.plot(kind='bar', color='darkorchid', edgecolor='black')
plt.title('Promedio de tiempo de servicio por área')
plt.xlabel('Área')
plt.ylabel('Promedio de tiempo de servicio (años)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
for i, v in enumerate(promedio_por_area):
    ax.text(i, v + 0.1, f'{v:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
data['Fecha fin'] = pd.to_datetime(data['Fecha fin'])
data['Año'] = data['Fecha fin'].dt.year
total_desvinculaciones_por_año = data.groupby('Año').size()
promedio_desvinculaciones_por_año = total_desvinculaciones_por_año.mean()
print("Promedio de desvinculaciones por año:", promedio_desvinculaciones_por_año)
```

```python
import matplotlib.pyplot as plt
data['Fecha fin'] = pd.to_datetime(data['Fecha fin'])

data['Año'] = data['Fecha fin'].dt.year
data['Mes'] = data['Fecha fin'].dt.month
desvinculaciones_por_area_y_mes = data.groupby(['Area', 'Año', 'Mes']).size().reset_index()
desvinculaciones_por_area_y_mes.columns = ['Área', 'Año', 'Mes', 'Cantidad de Empleados Desvinculados']
desvinculaciones_por_area = desvinculaciones_por_area_y_mes.groupby(['Área', 'Año'])['Cantidad de Empleados Desvinculados'].sum().reset_index()
plt.figure(figsize=(12, 6))
for area in desvinculaciones_por_area['Área'].unique():
    area_data = desvinculaciones_por_area[desvinculaciones_por_area['Área'] == area]
    plt.bar(area_data['Año'].astype(str),
            area_data['Cantidad de Empleados Desvinculados'],
            label=area)

plt.title('Cantidad de Empleados Desvinculados por Área y Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Empleados Desvinculados')
plt.legend()
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

# 🔴 RATIO DESVINCULACIÓN

```python
#Ratio
#desvinculacion x area
desvinculaciones_por_area = data['Area'].value_counts()
# Contar el total de desvinculaciones
total_desvinculaciones = len(data)
# Calcular el ratio de desvinculaciones por área
ratio_desvinculaciones_por_area = desvinculaciones_por_area / total_desvinculaciones
print("Ratio de desvinculaciones por área:")
print(ratio_desvinculaciones_por_area)
```

El ratios nos brindan una comprensión de la distribución de desvinculaciones en cada área de la organización. Por ejemplo, áreas con ratios más altos pueden requerir una atención especial en términos de retención de empleados, gestión de recursos humanos y desarrollo de estrategias para mejorar la satisfacción y el compromiso del personal.
IT (Tecnologías de la Información): El ratio de desvinculaciones en el área de IT es del 26.67%. Esto significa que aproximadamente el 26.67% de todas las desvinculaciones ocurrieron en el área de Tecnologías de la Información.

```python
ratio_desvinculaciones = {
    'IT': 0.266667,
    'Ventas': 0.200000,
    'Compras': 0.166667,
    'Producción': 0.133333,
    'Marketing': 0.133333,
    'Auditoria': 0.066667,
    'Legales': 0.033333
}


empleados_capacitados_por_area = {
    'IT': 50,
    'Ventas': 30,
    'Compras': 20,
    'Producción': 40,
    'Marketing': 35,
    'Auditoria': 10,
    'Legales': 5
}

df_ratio_desvinculaciones = pd.DataFrame(list(ratio_desvinculaciones.items()), columns=['Área', 'Ratio'])
df_empleados_capacitados = pd.DataFrame(list(empleados_capacitados_por_area.items()), columns=['Área', 'Empleados Capacitados'])

df_merged = pd.merge(df_ratio_desvinculaciones, df_empleados_capacitados, on='Área')

total_empleados = df_merged['Empleados Capacitados'].sum()
df_merged = df_merged.sort_values(by='Ratio')

plt.figure(figsize=(10, 6))

index = range(len(df_merged['Área']))
bar_width = 0.4

bar1 = plt.bar(index, df_merged['Empleados Capacitados'], bar_width, color='mediumpurple', label='Empleados Capacitados')
bar2 = plt.bar(index, df_merged['Empleados Capacitados'] - df_merged['Empleados Capacitados'] * df_merged['Ratio'],
               bar_width, bottom=df_merged['Empleados Capacitados'], color='mediumorchid', label='Empleados No Capacitados')

plt.text(bar1[-1].get_x() + bar1[-1].get_width() / 2, bar1[-1].get_height(), f'Total: {total_empleados}', ha='center', va='bottom')

plt.plot(index, df_merged['Ratio'] * df_merged['Empleados Capacitados'], color='black', marker='o', linestyle='-',
         linewidth=2, markersize=6, label='Ratio Desvinculaciones')

plt.title('Total de Empleados Capacitados y No Capacitados y Ratio de Desvinculaciones por Área')
plt.xlabel('Área')
plt.ylabel('Cantidad')
plt.xticks(index, df_merged['Área'], rotation=45, ha='right')
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
empleados_por_area = {
    'IT': 100,
    'Ventas': 80,
    'Compras': 60,
    'Producción': 120,
    'Marketing': 90,
    'Auditoria': 30,
    'Legales': 20
}

empleados_capacitados_por_area = {
    'IT': 50,
    'Ventas': 30,
    'Compras': 40,
    'Producción': 80,
    'Marketing': 60,
    'Auditoria': 20,
    'Legales': 10
}

data = []
for area, total_empleados in empleados_por_area.items():
    total_capacitados = empleados_capacitados_por_area.get(area, 0)
    total_no_capacitados = total_empleados - total_capacitados
    data.append([area, total_empleados, total_capacitados, total_no_capacitados])

df = pd.DataFrame(data, columns=['Área', 'Total Empleados', 'Empleados Capacitados', 'Empleados No Capacitados'])

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')
tabla = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='upper left')
tabla.auto_set_font_size(False)
tabla.set_fontsize(10)
tabla.scale(1.2, 1.2)

plt.title('Tabla de Total de Empleados por Área')
plt.show()
```

```python
empleados_por_area = data['Area'].value_counts().reset_index()
empleados_por_area.columns = ['Área', 'Cantidad de Empleados']
managers_por_area = data.groupby('Area')['Manager'].nunique().reset_index()
managers_por_area.columns = ['Área', 'Cantidad de Managers']
tabla = pd.merge(empleados_por_area, managers_por_area, on='Área', how='outer')
tabla = tabla.sort_values(by='Cantidad de Empleados', ascending=False)
plt.figure(figsize=(10, 8))
plt.bar(tabla['Área'], tabla['Cantidad de Empleados'], color='mediumpurple', label='Cantidad de Empleados')
plt.bar(tabla['Área'], tabla['Cantidad de Managers'], color='mediumorchid', label='Cantidad de Managers', alpha=0.7)

for index, value in enumerate(tabla['Cantidad de Empleados']):
    plt.text(index, value, str(value), ha='center', va='bottom', color='black')

for index, value in enumerate(tabla['Cantidad de Managers']):
    plt.text(index, value, str(value), ha='center', va='bottom', color='black')

plt.title('Cantidad de Empleados y Managers por Área')
plt.xlabel('Área')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.show()
```
